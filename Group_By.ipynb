```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'A': ['foo', 'bar', 'foo', 'bar', 'bar', 'foo', 'foo'],
    'B': [False, True, False, True, True, True, True],
    'C': [2.1, 1.9, 3.6, 4.0, 1.9, 7.8, 2.8],
    'D': [50, np.nan, 30, 90, 10, np.nan, 10]
}).convert_dtypes()
print(df,'\n')

s = pd.Series(
    data=[200., 100., 250., 50., 40., 60.],
    index= ['falcon', 'falcon', 'falcon', 'parrot', 'parrot', 'parrot']
    )
print(s)
```

```python
arrays = [['Falcon', 'Falcon', 'Parrot', 'Parrot'], ['Captive', 'Wild', 'Captive', 'Wild']]
index = pd.MultiIndex.from_arrays(arrays, names=('Animal', 'Type'))
s2 = pd.Series([200., 100., 50., 40.],
               index = index)
print(s2)
```

```python
s2.groupby(level=0).mean()
```

```python
s2.groupby(level='Type').mean()
```

```python
s3 = pd.Series([1, 2, 3, 3], index=["a", 'a', 'b', np.nan])
print(s3)
```

```python
s3.groupby(level=0, dropna=False).sum()
```

```python
df.groupby('A')[['C','D']].sum()
```

```python
df.groupby('A')[['C','D']].mean()
```

```python
df.groupby('A')[['C','D']].agg('sum')
```

```python
df.groupby('A').agg({'C': np.sum, 'D': ['mean', 'var']})
```

```python
df.groupby(by=['A', 'B'])[['D']].mean()
```

```python
df.groupby(by=['B', 'D']).agg({'C': ['min', 'max']})
```

```python
df.groupby(by=['B', 'D'], dropna=False).agg({'C': ['min', 'max']})
```

```python
df.groupby(by=['B','D'], dropna=False).agg(
    C_min=('C', 'min'),
    C_max=('C', np.max)
)
```
