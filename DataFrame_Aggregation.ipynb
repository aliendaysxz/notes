```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'x': [3.1, 5.5, 9.2, 1.7, 1.2, 8.3, 2.6],
    'y': [1.4, np.nan, 5.0, 5.8, 9.0, np.nan, 9.2]
})
print(df)
```

```python
df.agg(sum)
```

```python
df.agg(np.sum)
```

```python
df.agg(sum, axis=1)
```

```python
df.agg('sum')
```

```python
df.agg(['sum'])
```

```python
df.agg(['sum', 'mean'])
```

```python
df.agg(np.sort)
```

```python
def foofunc(x):
  return x.iloc[1]
```

```python
df.agg(foofunc)
```

```python
df.agg({
    'x': ['sum','mean'],
    'y': ['min', 'max']
})
```

```python
df.agg(['sum','mean'],axis=1)
```

```python
df.agg('min')
```

```python
df.agg('min',axis=1)
```

```python
df.agg(['argmin'], axis=1)
```

```python
df.agg(['argmin'])
```

```python
def nth_value(p, n1, n2):
  return p.iloc[n1] + p.iloc[n2]
```

```python
df.agg(nth_value, n1=4, n2=1)
```

```python
df.agg(lambda p, n: p.iloc[n], n=4)
```

```python
df.agg(lambda p, n: p.iloc[n], n=4)
```

```python
df.agg(lambda p, n1, n2: p.iloc[n1] + p.iloc[n2], n1=2, n2=6)
```
